```python
import pandas as pd
import numpy as np

df=pd.read_csv('C:/Users/User/Documents/Strive_AI_Jun_21/M4-Feature Engineering/03. Data Enhancement/data/london_merged.csv')
# print(df)
df.head()
```

```python
X,y = df.drop(['cnt'],axis=1),df['cnt']
```

```python
def data_enhancement(df):
    
    df_ehn = df.copy()
    
    for ssn in df['season'].unique():
        ssn_data =  df_ehn[df_ehn['season'] == ssn]
        hum_std = ssn_data['hum'].std()/10
        wind_speed_std = ssn_data['wind_speed'].std()/10
        t1_std = ssn_data['t1'].std()/10
        t2_std = ssn_data['t2'].std()/10
        
        for i in df_ehn[df_ehn['season'] == ssn].index:
            if np.random.randint(2) == 1:
                df_ehn['hum'].values[i] += hum_std
            else:
                df_ehn['hum'].values[i] -= hum_std
                
            if np.random.randint(2) == 1:
                df_ehn['wind_speed'].values[i] += wind_speed_std
            else:
                df_ehn['wind_speed'].values[i] -= wind_speed_std
                
            if np.random.randint(2) == 1:
                df_ehn['t1'].values[i] += t1_std
            else:
                df_ehn['t1'].values[i] -= t1_std
                
            if np.random.randint(2) == 1:
                df_ehn['t2'].values[i] += t2_std
            else:
                df_ehn['t2'].values[i] -= t2_std

    return df_ehn
```
