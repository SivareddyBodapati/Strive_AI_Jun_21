### To edit an image making a copy of it is always a good idea!

```python
# Load an image
# And make a copy of the original image to avoid manipulating the original image
img = cv2.imread('img/blue-flowers.jpg')
img_copy = img.copy()
img1 = cv2.imread('img/blue-red-flowers.png')
img1_copy = img1.copy()
img2 =cv2.imread('img/hsv.png')
img2_copy = img2.copy()
img3 = cv2.imread('img/m.jpg')
img3_copy = img3.copy()
img4 = cv2.imread('img/purple-flowers.jpg')
img4_copy = img4.copy()
```

```python
green =(0,255,0)
red=(255,0,0)
blue =(0,0,255)
```

[Hint](# "Numpy .copy function") 

## Drawing

### Drawing lines

```python
plt.figure(figsize=(10,15))
plt.imshow(img)
print(img.shape)
```

```python
# Draw some lines on top of the image
cv2.line(img_copy,(10,20),(250,140),green,3)
plt.imshow(img_copy)
```

[Hint](# "cv2.line function") 

### Circles

```python
# Draw some circles on top of the image
img_copy = img.copy()
cv2.circle(img_copy,(80,20),20,red,cv2.FILLED)
plt.figure(figsize=(10,15))
plt.imshow(img_copy)
```

[Hint](# "cv2.circle") 

### Ellipses

```python
# Draw some elipses on top of the image
img_copy = img.copy()
cv2.ellipse(img_copy,(140,80),(80,40),45,0,360,blue,-1)
plt.imshow(img_copy)
```

[Hint](# "cv2.ellipse") 

### Poligons

```python
# Lets draw some poligons, first declare the points of the poligon
pts =np.array([[10,5],[120,30],[170,120],[250,140]])
```

[Hint](# "Example: np.array([[10,5],[120,130],[170,120],[250,210]], np.int32)") 

```python
# Draw the poligon on top of the image
img_copy=img.copy()
cv2.polylines(img_copy,[pts],True,green,3) # False can be use din place of ture to not close the polygon
plt.imshow(img_copy)
```

### Add an overlay on top of an image

```python
# Load an image and create a copy of it you can call it overlay
# Create a rectangle with a fill color (-1) on top of the overlay image
# Combine both images setting some opacity on the overlay
# Add some text on top of the overlay and show the image on screen
```

[Hint](# "cv2.addWeighted")  

```python
img_copy=img.copy() 
overlay = img.copy() # for an overlay image

cv2.rectangle(overlay,(100,20),(200,60),(255,255,255),-1)
cv2.putText(overlay,'SIVA',(120,50),cv2.FONT_HERSHEY_COMPLEX, 1, green, 2)
opacity = 0.8
beta =1-opacity
image = cv2.addWeighted(overlay,opacity,img_copy,beta,10)

plt.imshow(image)
```
