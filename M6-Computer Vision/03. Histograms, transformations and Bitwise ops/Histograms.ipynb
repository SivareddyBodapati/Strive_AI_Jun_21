```python
def imshow(img, grayscale=False):
    plt.figure(figsize=(10,15))
    if grayscale:
        plt.imshow(img, cmap='gray') # black image
    else:
        plt.imshow(img[:,:,::-1]) # reverse order form BGR to RGB
```

```python
imshow(img_night4)
```

```python
blue_hist = cv2.calcHist([img_night4],[0],None,[255],[0,255]) # histogram of the image  (img,channel (BGR),mask, histsize, range of histgram)
red_hist = cv2.calcHist([img_night4],[2],None,[255],[0,255])
green_hist = cv2.calcHist([img_night4],[1],None,[255],[0,255])
```

```python
blue_hist # pixels
```

```python
plt.plot(blue_hist,c='b')
```

```python
plt.plot(green_hist, c='g')
```

```python
plt.plot(red_hist, c='r')
```
